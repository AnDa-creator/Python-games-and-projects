```python
from qiskit import *
from math import pi
from qiskit.visualization import plot_bloch_multivector,plot_histogram, plot_bloch_vector

def y_measurement(qc, qubit, cbit):
    """Measure 'qubit' in the Y-basis, and store the result in 'cbit'"""
    qc.h(qubit)
    qc.s(qubit)
    qc.measure(qubit, cbit)
    qc.sdg(qubit)
    qc.h(qubit)
    return qc

q_c = QuantumCircuit(1, 1)
q_c = y_measurement(q_c, 0, 0)
q_c.draw(output="mpl")
```

```python
backend = Aer.get_backend('statevector_simulator') # Tell Qiskit how to simulate our circuit
out_state1 = execute(q_c,backend).result().get_statevector() # Do the simulation, returning the state vector
plot_bloch_multivector(out_state1) # Display the output state vector
```
